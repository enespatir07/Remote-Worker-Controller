```python
from ultralytics import YOLO
import cv2
from matplotlib import pyplot as plt
import shutil

# Load the trained YOLO model
model = YOLO('runs/train/drowsy_detection2/weights/best.pt')

def capture_and_predict():
    # Open the camera
    cap = cv2.VideoCapture(0)  # Use 0 for the default camera
    if not cap.isOpened():
        print("Error: Could not open camera.")
        return

    print("Press 's' to capture an image.")
    while True:
        # Capture frame-by-frame
        ret, frame = cap.read()
        if not ret:
            print("Error: Could not read frame.")
            break

        # Display the live video feed
        cv2.imshow('Camera', frame)

        # Wait for 's' key to capture the image
        if cv2.waitKey(1) & 0xFF == ord('s'):
            captured_image = frame
            break

    # Release the camera and close the window
    cap.release()
    cv2.destroyAllWindows()

    # Save the captured image to a temporary file
    temp_image_path = 'captured_image.jpg'
    cv2.imwrite(temp_image_path, captured_image)
    print(f"Image captured and saved to {temp_image_path}")

    # Perform prediction
    results = model.predict(source=temp_image_path)  # Perform prediction

    # Display predictions
    detections = results[0].boxes
    print("Detections:")
    for box in detections:
        cls_id = int(box.cls[0])  # Class ID
        conf = box.conf[0]        # Confidence score
        x1, y1, x2, y2 = box.xyxy[0].tolist()  # Bounding box coordinates
        print(f"Class ID: {cls_id}, Confidence: {conf:.2f}, BBox: [{x1:.2f}, {y1:.2f}, {x2:.2f}, {y2:.2f}]")

    # Visualize the results
    annotated_frame = results[0].plot()  # Annotated image with bounding boxes
    plt.imshow(cv2.cvtColor(annotated_frame, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()
    return temp_image_path

def move_image(input_image_path, user_input):
    # Hedef klasörleri tanımla
    folders = {
        '0': 'new_Birden fazla',
        '1': 'new_Bulamadı',
        '2': 'new_Buldu'
    }
    
    if user_input not in folders:
        print("Geçersiz değer! Lütfen 0, 1 veya 2 girin.")
        return
    
    target_folder = folders[user_input]
    
    # Hedef klasör yoksa oluştur
    if not os.path.exists(target_folder):
        os.makedirs(target_folder)
    
    # Hedef klasördeki mevcut dosya sayısını al
    file_count = len([name for name in os.listdir(target_folder) if os.path.isfile(os.path.join(target_folder, name))])
    new_image_name = f"{file_count + 1}.png"
    target_path = os.path.join(target_folder, new_image_name)
    
    # Görüntüyü taşı ve yeniden adlandır
    shutil.move(input_image_path, target_path)
    print(f"Görüntü {target_folder} klasörüne taşındı ve adı {new_image_name} olarak değiştirildi.")

# Run the function
input_image_path = capture_and_predict()
user_input = input("Görüntü sonucu (0: Birden fazla, 1: Bulamadı, 2: Buldu): ").strip()
move_image(input_image_path, user_input)
```

```python

from ultralytics import YOLO
import os
import cv2
from matplotlib import pyplot as plt
import shutil

model = YOLO('runs/train/bitirme/weights/best.pt')

# Test the trained YOLO model
def test_model(input_image_path):
    # Read the input image
    image = cv2.imread(input_image_path)
    
    # Convert the image from BGR (OpenCV) to RGB (matplotlib)
    rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    # Run the model on the image
    results = model.predict(source=input_image_path, show=False, save=False)

    # Plot the results using matplotlib
    plt.figure(figsize=(10, 10))
    plt.imshow(rgb_image)
    
    # Draw detections on the image
    for box in results[0].boxes.data:
        x1, y1, x2, y2, conf, cls = box.tolist()
        plt.gca().add_patch(plt.Rectangle((x1, y1), x2 - x1, y2 - y1, fill=False, edgecolor='red', linewidth=2))
        plt.text(x1, y1 - 10, f"{model.names[int(cls)]} {conf:.2f}", color='red', fontsize=12, backgroundcolor='white')

    plt.axis('off')
    plt.show()

def move_image(input_image_path, user_input):
    # Hedef klasörleri tanımla
    folders = {
        '1': 'found2',
        '2': 'found_not2',
    }
    
    if user_input not in folders:
        print("Geçersiz değer! Lütfen 0, 1 veya 2 girin.")
        return
    
    target_folder = folders[user_input]
    
    # Hedef klasör yoksa oluştur
    if not os.path.exists(target_folder):
        os.makedirs(target_folder)
    
    # Hedef klasördeki mevcut dosya sayısını al
    file_count = len([name for name in os.listdir(target_folder) if os.path.isfile(os.path.join(target_folder, name))])
    new_image_name = f"{file_count + 1}.png"
    target_path = os.path.join(target_folder, new_image_name)
    
    # Görüntüyü taşı ve yeniden adlandır
    shutil.move(input_image_path, target_path)
    print(f"Görüntü {target_folder} klasörüne taşındı ve adı {new_image_name} olarak değiştirildi.")

image_directory = './'
supported_extensions = ['.jpg', '.png', '.jpeg']

# Tüm görüntü dosyalarını işle
for file_name in os.listdir(image_directory):
    # Dosya uzantısını kontrol et
    if any(file_name.lower().endswith(ext) for ext in supported_extensions):
        input_image_path = os.path.join(image_directory, file_name)

        print(f"\nGörüntü işleniyor: {file_name}")
        test_model(input_image_path)

        user_input = input("Görüntü sonucu (1: Buldu, 2: Bulamadı): ").strip()
        move_image(input_image_path, user_input)
```

```python
import cv2
from ultralytics import YOLO

model = YOLO('runs/train/bitirme/weights/best.pt')  # Eğitimli modelinizi yükleyin

def process_video(video_path=None):
    cap = cv2.VideoCapture(0 if video_path is None else video_path)
    
    if not cap.isOpened():
        print("Kamera veya video açılmadı.")
        return

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            print("Video akışı sonlandı veya hata oluştu.")
            break

        results = model.predict(source=frame, show=False)  # Kare üzerinde tahmin yap

        detections = results[0].boxes.data
        for box in detections:
            x1, y1, x2, y2, conf, cls = box.tolist()
            cls_name = model.names[int(cls)]
            print(f"Algılandı: {cls_name}, Güven: {conf:.2f}, Kutu: [{x1}, {y1}, {x2}, {y2}]")

        # Tespit edilen kutuları göster
        annotated_frame = results[0].plot()
        cv2.imshow('Video Akışı', annotated_frame)

        if cv2.waitKey(1) & 0xFF == ord('q'):  # 'q' tuşuna basılarak çıkılabilir
            break

    cap.release()
    cv2.destroyAllWindows()

process_video()  # Kameradan algılamayı başlat
```
